```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Perceptron

# CSV Load karo
df = pd.read_csv("/content/Placement.csv")

# Scatter Plot
plt.figure(figsize=(4,3))
sns.scatterplot(x="CGPA", y="IQ", data=df, hue="Placement")
plt.show()

# Sirf CGPA aur IQ wale features select karo
x = df[['CGPA', 'IQ']]
y = df["Placement"]

# Train-test split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=10)

# Perceptron Model Train karo
pr = Perceptron()
pr.fit(x_train, y_train)

# Model ka Score
print("Test Score:", pr.score(x_test, y_test) * 100)
print("Train Score:", pr.score(x_train, y_train) * 100)

# Decision Boundary Plot
plot_decision_regions(x.to_numpy(), y.to_numpy(), clf=pr)
plt.xlabel("CGPA")
plt.ylabel("IQ")
plt.show()
```
